```python
import matplotlib.pyplot as plt
import numpy as np

# Example data: temperature in Celsius for a month
temperatures = np.array([15, 18, 20, 22, 23, 25, 27, 28, 26, 24, 22, 20, 18, 16, 14, 12, 10, 12, 14, 16, 18, 20, 22, 24, 25, 23])

# Set the number of bins (adjust as needed)
bins = 10

# Create the histogram
plt.hist(temperatures, bins=bins, edgecolor='black')

# Label the axes
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')

# Add a title
plt.title('Distribution of Daily Temperatures in May')

# Show the plot
plt.grid(axis='y')  # Add gridlines for better readability
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data: temperature in Celsius for a month
temperatures = np.array([15, 18, 20, 22, 23, 25, 27, 28, 26, 24, 22, 20, 18, 16, 14, 12, 10, 12, 14, 16, 18, 20, 22, 24, 25, 23])
death = np.random.randint(5, 50, size=len(temperatures))
print(temperatures)
print(death)

# Set the number of bins (adjust as needed)
bins = 10

# Create the histogram
plt.hist2d(temperatures, death, bins=bins)

# Label the axes
plt.xlabel('Temperature (°C)')
plt.ylabel('Death')

# Add a title
plt.title('Distribution of Daily Temperatures in May')

plt.colorbar()  # Add a colorbar for better interpretation
plt.show()
```
