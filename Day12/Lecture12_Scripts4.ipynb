# Visualizing Errors

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python

data = {'Model1': [10,12,32,45,67,72],'Error': [2,5,2,12,6,10]}
df = pd.DataFrame (data, columns = ['Model1','Error'], index=['I1', 'I2','I3','I4','I5','I6'])
df
```

```python
x = df["Model1"]
y = df['Error']

plt.errorbar(df.index, x, yerr=y, fmt='-ok');
```

```python
x = df["Model1"]
y = df['Error']

plt.errorbar(df.index, x, yerr=y, fmt='.k');
```

```python
plt.errorbar(df.index, x, yerr=y, fmt='o', color='black',
                        ecolor='lightgray', elinewidth=3, capsize=5);
```
